```python
import urllib.request
```

```python
def read_html():  
    user = 'Google Chrome/61.0 (Windows NT 6.1; Win64; x64)'
    page = urllib.request.Request('https://github.com/dashapopova/Slovo-dnja/blob/master/Entries.md', headers={'User-Agent':user})
    with urllib.request.urlopen(page) as p:
        html = p.read().decode('utf-8')
    return html
```

```python
import re
```

```python
def del_space():
    reg_space = re.compile('\s{2,}', re.DOTALL)
    html = read_html()
    html = reg_space.sub(" ", html)
    return html
```

```python
reg_info = re.compile('<\/svg><\/a>(.*?)<p>...<\/p>', re.DOTALL)
reg_topic = re.compile('<strong>Тема:<\/strong>(.*?)<\/p>', re.DOTALL)
reg_expl = re.compile('<p><strong>Объяснение:<\/strong>(.*?)<strong>', re.DOTALL)
reg_examp = re.compile('<strong>Примеры:(.*?)(<ol|<\/p>\s+[А-Яа-я]|<\/p>\s$)', re.DOTALL)
```

```python
html = del_space()
infos = re.findall(reg_info, html)
```

```python
topics = []
reg_tag = re.compile('<.*?>', re.DOTALL)
for i in infos:
    if re.search(reg_topic, i):
        topic = re.findall(reg_topic, i)
        for t in topic:
            t = reg_tag.sub("", t)
            topics.append(t)
topics
```

```python
expls = []
reg_tag = re.compile('<.*?>', re.DOTALL)
reg_n = re.compile('\n', re.DOTALL)
for i in infos:
    if re.search(reg_expl, i):
        expl = re.findall(reg_expl, i)
        for e in expl:
            e = reg_tag.sub("", e)
            e = reg_n.sub("", e)
            expls.append(e)
expls
```

```python
examps = []
reg_tag = re.compile('<.*?>', re.DOTALL)
reg_n = re.compile('\n', re.DOTALL)
reg_space = re.compile('\s', re.DOTALL)

for i in infos:
    if re.search(reg_examp, i):
        i = reg_space.sub(' ', i)
        examp = re.findall(reg_examp, i)
        for ex in examp:
            e = ex[0]
            e = reg_tag.sub("", e)
            e = reg_n.sub("", e)
            examps.append(e)        
examps
```

```python
table = [(i,j,k) for i,j,k in zip(topics,expls,examps)]
table
```

```python
import sqlite3

connection = sqlite3.connect("database.db")

cursor = connection.cursor()

cursor.execute("""DROP TABLE russian;""")

sql_command = """
CREATE TABLE russian ( 
Topics,
Explanations,
Examples);"""

cursor.execute(sql_command)

for t in table:
    format_str = """INSERT INTO russian (Topics, Explanations, Examples)
    VALUES ('{topic}', '{expl}', '{examp}');"""
    sql_command = format_str.format(topic=t[0], expl=t[1], examp=t[2])
    cursor.execute(sql_command)

connection.commit()

connection.close()
```
